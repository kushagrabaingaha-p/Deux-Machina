```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('extended_salary_data.csv')
year=df['YearsExperience']
sal=df['Salary']
year_mean=year.mean()
sal_mean=sal.mean()
year_var=year.var()
sal_var=sal.var()
year_med=year.median()
sal_med=sal.median()

print(year_mean,year_med,year_var,sal_mean,sal_var,sal_med)
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],c='r')
plt.xlabel('Year Of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
X = df['YearsExperience'].to_numpy().reshape(-1, 1)
y = df['Salary'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
sgdr=SGDRegressor(max_iter=1000)

sgdr.fit(X_train, y_train)

intercept=sgdr.intercept_
slope=sgdr.coef_
y_pred=sgdr.predict(X_test)
```

```python
plt.scatter(X_test,y_test,c='r')
plt.plot(X_test,y_pred,c='b')
plt.xlabel('Year Of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
error_sq_sum=(y_pred-y_test)**2
error=y_pred-y_test
error_abs=np.abs(error)

MSE=(error_sq_sum.sum())/error_sq_sum.shape[0]
MAE=error_abs.sum()/error_abs[0]
```
